### AUC-ROC Plot: Treated fracture Detection on Test Dataset: (A) Upper Limb (B) Lower Limb

```python
body_parts = {
    'forearm': 'forearm_target',
    'foot': 'foot_target',
    'humerus': 'humerus_target',
    'pelvis': 'pelvis_target',
    'leg': 'leg_target',
    'ankle': 'ankle_target',
    'clavicle': 'clavicle_target',
    'knee': 'knee_target',
    'fibula': 'fibula tibia_target',
    'femur': 'femur_target',
    'hip': 'hip_target',
    'wrist': 'wrist_target',
    'elbow': 'elbow_target',
    'hand': 'hand_target',
    'finger': 'finger_target',
    'shoulder': 'shoulder_target',
    'toe': 'toe_target'
}

fpr_treated = {}
tpr_treated = {}

for part, target in body_parts.items():
    intern_df = internal_df[internal_df[target] == 1].reset_index(drop=True)
    fpr, tpr, _ = metrics.roc_curve(intern_df['treated_fracture_target'], intern_df['treated_fracture'])
    fpr_treated[part] = fpr
    tpr_treated[part] = tpr

fpr_treated['overall'], tpr_treated['overall'], _ = metrics.roc_curve(
    internal_df['treated_fracture_target'], internal_df['treated_fracture']
)

fig, ax2 = plt.subplots(1, 1, figsize=(10, 6))
colors = ['red', 'green', 'blue', 'olive', 'coral', 'limegreen', 'orange', 'indigo']
markers = ['^', '^', '^', '*', '^', '^', '^', '^']
labels = ['Clavicle', 'Shoulder', 'Humerus', 'Elbow', 'Forearm', 'Wrist', 'Hand', 'Finger']

for part, color, marker, label in zip(
    ['clavicle', 'shoulder', 'humerus', 'elbow', 'forearm', 'wrist', 'hand', 'finger'],
    colors, markers, labels
):
    ax2.plot(fpr_treated[part], tpr_treated[part], color=color, marker=marker,
             linestyle='dashdot', markersize=3, label=label)

ax2.grid(True)
ax2.legend(loc='lower right', fontsize=20)
ax2.set_title('(A)', fontsize=25)
ax2.set_ylabel('Sensitivity', fontsize=23)
ax2.set_xlabel('1 - Specificity', fontsize=23)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.show()

fig, ax3 = plt.subplots(1, 1, figsize=(10, 6))
colors = ['darkviolet', 'indigo', 'crimson', 'olivedrab', 'lime', 'blue', 'teal', 'magenta', 'salmon']
labels = ['Pelvis', 'Hip', 'Femur', 'Knee', 'Tibia and Fibula', 'Leg', 'Ankle', 'Foot', 'Toe']

for part, color, label in zip(
    ['pelvis', 'hip', 'femur', 'knee', 'fibula', 'leg', 'ankle', 'foot', 'toe'],
    colors, labels
):
    ax3.plot(fpr_treated[part], tpr_treated[part], color=color, marker='*',
             linestyle='-', markersize=2, label=label)

ax3.grid(True)
ax3.legend(loc='lower right', fontsize=20)
ax3.set_title('(B)', fontsize=25)
ax3.set_xlabel('1 - Specificity', fontsize=23)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.show()

# fig, ax4 = plt.subplots(1, 1, figsize=(10, 6))
# ax4.plot(fpr_treated['overall'], tpr_treated['overall'], 'blue', marker='.', 
#          linestyle='dashdot', markersize=3, label='Overall')
# ax4.grid(True)
# ax4.legend(loc='lower right', fontsize=20)
# ax4.set_title('(C)', fontsize=25)
# ax4.set_xlabel('1 - Specificity', fontsize=23)
# plt.xticks(fontsize=20)
# plt.yticks(fontsize=20)
# plt.show()
```

# Erasmus University Medical Center dataset

###  AUC-ROC Plot: Fracture Detection on Erasmus University Medical Center dataset: (A) Upper Limb (B) Lower Limb

```python
erasmus_df = pd.read_csv('Dataset/External/Erasmus_Final_CSV.csv')

body_parts = {
    'Pelvis': 'pelvis',
    'Femur': 'femur',
    'Hip': 'hip',
    'Wrist': 'wrist',
    'Fibula and Tibia': 'fibula_tibia'
}

fpr = {}
tpr = {}
auc = {}

dfs_to_concat = []

for part, label in body_parts.items():
    part_df = erasmus_df[erasmus_df['Final BP'] == part].reset_index(drop=True)
    fpr[label], tpr[label], _ = metrics.roc_curve(part_df['fracture_gt'], part_df['fracture_score'])
    auc[label] = metrics.roc_auc_score(part_df['fracture_gt'], part_df['fracture_score'])
    dfs_to_concat.append(part_df)

s = pd.concat(dfs_to_concat).reset_index(drop=True)

fpr_overall, tpr_overall, _ = metrics.roc_curve(s['fracture_gt'], s['fracture_score'])
auc_overall = metrics.roc_auc_score(s['fracture_gt'], s['fracture_score'])

fig, ax2 = plt.subplots(1, 1, figsize=(9, 6))
colors = ['orangered', 'limegreen', 'indigo', 'magenta', 'teal']
markers = ['*', '+', '*', '.', '+']
labels = ['Pelvis', 'Hip', 'Femur', 'Tibia and Fibula', 'Wrist']

for label, color, marker, plot_label in zip(body_parts.values(), colors, markers, labels):
    ax2.plot(fpr[label], tpr[label], color=color, marker=marker, linestyle='dashdot', label=plot_label)

ax2.grid(True)
ax2.legend(loc='lower right', fontsize=15)
ax2.set_title('(A)', fontsize=25)
ax2.set_xlabel('1-Specificity', fontsize=22)
ax2.set_ylabel('Sensitivity', fontsize=22)

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
fig.tight_layout()
# plt.savefig('../../Erasmus_plot3_1.png', dpi=400)
plt.show()

fig, ax3 = plt.subplots(1, 1, figsize=(9, 6))
ax3.plot(fpr_overall, tpr_overall, 'blue', marker='*', linestyle='dashdot', label='Overall')

ax3.grid(True)
ax3.legend(loc='lower right', fontsize=18)
ax3.set_title('(B)', fontsize=25)
ax3.set_xlabel('1-Specificity', fontsize=18)

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
# plt.savefig('../../testing_internal_data_overall.png', dpi=400)
plt.show()
```

# Open-source Validation Datasets

### AUC-ROC Plot: Fracture Detection on Open-source Validation Datasets: (A) FracAtlas (B) GRAZPEDWRI-DX

```python
frac_atlas = pd.read_csv('Dataset/Open-source/frac_atlas_results.csv')
paediatric_dataset = pd.read_csv('Dataset/Open-source/paediatric_results.csv')

body_parts = ['leg', 'hand', 'hip', 'shoulder']
filtered_dfs = {}

for part in body_parts:
    filtered_dfs[f'{part}_df'] = frac_atlas[frac_atlas[part] == 1].reset_index(drop=True)

datasets = {
    'Leg': filtered_dfs['leg_df'],
    'Hand': filtered_dfs['hand_df'],
    'Hip': filtered_dfs['hip_df'],
    'Shoulder': filtered_dfs['shoulder_df'],
    'Overall': frac_atlas,
}

fpr_atlas = {}
tpr_atlas = {}

for label, df in datasets.items():
    fpr_atlas[label], tpr_atlas[label], _ = metrics.roc_curve(df['fractured'], df['fracture_average'])

fig, ax3 = plt.subplots(1, 1, figsize=(10, 6))
colors = ['darkviolet', 'lime', 'crimson', 'olivedrab', 'blue']
linestyles = ['-', 'dashdot', '-', '-', '-']
labels = ['Leg', 'Hand', 'Hip', 'Shoulder', 'Overall']

for label, color, linestyle in zip(datasets.keys(), colors, linestyles):
    ax3.plot(fpr_atlas[label], tpr_atlas[label], color=color, marker='*', linestyle=linestyle, label=label)

ax3.grid(True)
ax3.legend(loc='lower right', fontsize=20)
ax3.set_title('(A)', fontsize=25)
ax3.set_ylabel('Sensitivity', fontsize=21)
ax3.set_xlabel('1 - Specificity', fontsize=21)

plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.show()

fpr_paediatric, tpr_paediatric, _ = metrics.roc_curve(paediatric_dataset['fracture_visible'], paediatric_dataset['fracture_average'])

fig, ax3 = plt.subplots(1, 1, figsize=(10, 6))

ax3.plot(fpr_paediatric, tpr_paediatric, 'blue', marker='*', linestyle='-', label='Overall')
ax3.grid(True)
ax3.legend(loc='lower right', fontsize=20)
ax3.set_title('(B)', fontsize=25)
ax3.set_xlabel('1 - Specificity', fontsize=21)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.show()
```
